```python
#2. Data Analysis
#Q4: Check for missing values in the dataset.
```

```python
print(df.isnull().sum())
```

```python
#Q5: What is the average monthly income of employees?
```

```python
print(df['MonthlyIncome'].mean())
```

```python
#Q6: Find the percentage of employees who have left the company (Attrition = 'Yes').
```

```python
attrition_rate = (df['Attrition'].value_counts(normalize=True) * 100)['Yes']
print(attrition_rate)
```

```python
#3. Data Filtering and Manipulation
#Q7: Filter employees who are in the "Sales" department and have "OverTime" as "Yes."
```

```python
sales_overtime = df[(df['Department'] == 'Sales') & (df['OverTime'] == 'Yes')]
print(sales_overtime)
```

```python
#Q8: Create a new column that shows the annual income of employees.
```

```python
df['AnnualIncome'] = df['MonthlyIncome'] * 12
print(df.head())
```

```python
# Q9: Sort the dataset by "YearsAtCompany" in descending order.
```

```python
df_sorted = df.sort_values(by='YearsAtCompany', ascending=False)
print(df_sorted)
```

```python
#4. Data Visualization
#Q10: Plot the distribution of "MonthlyIncome."
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['MonthlyIncome'], kde=True, bins=30)
plt.title("Distribution of Monthly Income")
plt.show()
```

```python
#Q11: Create a bar chart to compare attrition counts by gender.
```

```python
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title("Attrition by Gender")
plt.show()
```

```python
#Q12: Create a scatter plot showing the relationship between "Age" and "MonthlyIncome."
```

```python
plt.scatter(df['Age'], df['MonthlyIncome'], alpha=0.5)
plt.title("Age vs Monthly Income")
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.show()
```

```python
#5. Statistical Analysis
#Q13: Calculate the correlation between "Age" and "MonthlyIncome."
```

```python
correlation = df['Age'].corr(df['MonthlyIncome'])
print("Correlation between Age and MonthlyIncome:", correlation)
```

```python
#Q14: Find the department with the highest average job satisfaction.
```

```python
department_satisfaction = df.groupby('Department')['JobSatisfaction'].mean()
print(department_satisfaction.idxmax())
```

```python
#Q15: Perform a t-test to compare the "MonthlyIncome" of employees with and without overtime.
```

```python
from scipy.stats import ttest_ind
overtime_income = df[df['OverTime'] == 'Yes']['MonthlyIncome']
no_overtime_income = df[df['OverTime'] == 'No']['MonthlyIncome']
t_stat, p_value = ttest_ind(overtime_income, no_overtime_income)
print("T-statistic:", t_stat, "P-value:", p_value)
```
